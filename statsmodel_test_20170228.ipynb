```python
# coding: utf-8
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 파일 로드
data = np.loadtxt("data.txt")
# 파일을 1차원 ndarray로 변환
x = data.T[0]
y = data.T[1]

# サンプルの数
nsample = x.size
print(nsample)
# おまじない (後で解説)
X = np.column_stack((np.repeat(1, nsample), x))

# 회귀실행
model = sm.OLS(y, X)
results = model.fit()

# 結果の概要を表示
print(results.summary())

# パラメータの推定値を取得
a, b = results.params

# プロットを表示
plt.plot(x, y, 'o')
plt.plot(x, a+b*x)
plt.text(0, 0, "a={:8.3f}, b={:8.3f}".format(a,b))
plt.show()
```

```python
import numpy as np
from itertools import product

# sample 만들기
# x 주사위 시행횟수
def dice(x) :
    a = np.arange(1,7)
    
    b = product(a,repeat=x)
    c = [ x for x in b]
    return c

#  event 추출
def event(iterable, ev) :
    d = []
    for x in iterable :
        if np.sum(x) == ev :
            d.append(x)
    return d
        
print(event(dice(2),7))
print(event(dice(3),7))
  
```

```python
num_firends = [100,40,30,30,30,30,30,30,30,30,54,54,54,54,54,54,54,54,25,3,100,100,100,3,3]

def mean(x) :
    return sum(x) / len(x)

avgOfFriends = mean(num_firends)
print(avgOfFriends)

def median(v) :
    n = len(v)
    sorted_v = sorted(v)  
    midpoint = n // 2 
    # 리스트가 홀 수면 가운데 값 
    if n % 2 == 1:
        return sorted_v[midpoint] # 리스트가 홀 수면 가운데 값 
    else :
        lo = midpoint - 1 
        hi = midpoint
        return (sorted_v[lo]+sorted_v[hi]) / 2

medianOfFriends = median(num_firends)
print(medianOfFriends)
```

```python
def fact(n) :
    if n == 0 or n == 1 : 
        return 1
    return n * fact(n-1)

print(fact(3))
print(fact(5))

def perm(n,r) :
    #분자
    numerator = fact(n)
    #분모
    denominator = fact(n-r)
    return numerator/denominator

print(perm(5,3))

def comb(n,r) :
    numerator = perm(n,r)
    denominator = fact(r)
    return numerator/denominator

print(comb(5,3))
```

```python
def dec_bin(x) :
    mod_val =[]
    
    #bin 함수의 결과는 문자열 0b이 길이를 뺌
    for i in range(len(bin(x))-2) :
        x,mod= divmod(x, 2)
        #문자열 연결을 위해 숫자를 문자로 변환
        mod_val.insert(0,str(mod))
    s = "0b" + "".join(mod_val)
    return s

print(dec_bin(6))
print(bin(6) == dec_bin(6))
```

```python
%%writefile hello.py
from flask import Flask, request
app = Flask(__name__)
print(app)

@app.route('/')
def index() :
    user_agent = request.headers.get('User-Agent')
    return "<h1> Hello World %s</h1>" % user_agent

@app.route('/user/<name>')
def user(name) :
    return "<h1> Hello World %s !</h1>" % name

if __name__ == "__main__" :
    app.run()
```

```python
from hello import app
from flask import current_app

app_ctx = app.app_context()
app_ctx.push()
print(current_app.name)
app.run()
```

```python
%%writefile app/route.py
from flask import Flask, render_template
  
app = Flask(__name__)
  
@app.route('/')
def home():
    return render_template('home.html')
  
@app.route('/about')
def about():
    #return 'The about page'
    return render_template('about.html')
  
if __name__ == '__main__':
    app.run(debug=True)
```
